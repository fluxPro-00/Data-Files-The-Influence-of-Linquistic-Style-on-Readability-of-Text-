```python
import pandas as pd

df = pd.read_excel("Concreteness_ratings_Brysbaert_et_al_BRM.xlsx")

df
```

```python
min_value = df["Conc.M"].min()
max_value = df["Conc.M"].max()

print(f"Minimum Conc.M: {min_value}")
print(f"Maximum Conc.M: {max_value}")
```

```python
columns_to_drop = ["Bigram", "Unknown", "Total", "Percent_known", "SUBTLEX", "Dom_Pos"]
df = df.drop(columns=columns_to_drop)
df
```

```python
df["Category"] = df["Conc.M"].apply(lambda x: "Concrete" if x >= 3 else "Abstract")

print(df.head())
```

```python
df.to_csv("categorized_words.csv", index=False, encoding="utf-8")
```
